```python
@nw.kernel
def program_1():
    srate: nw.Var = 4e9
    time_width_1: nw.Var = 200e-9
    freq_1: nw.Var = 50e6
    time_line_1: np.ndarray =np.linspace(0, time_width_1, int(time_width_1 * srate),endpoint=False)
    wave_1: np.ndarray = np.cos(2 * np.pi * time_line_1 * freq_1)

    time_width_2: nw.Var = 100e-9
    freq_2: nw.Var = 0e9
    time_line_2: np.ndarray =np.linspace(0, time_width_2, int(time_width_2 * srate),endpoint=False)
    wave_2: np.ndarray = np.cos(2 * np.pi * time_line_2 * freq_2)
    
    frame_0: nw.Frame = nw.init_frame(0, 0.5*np.pi)
    envelope_1: nw.Envelope = nw.ins_envelope(wave_1)
    envelope_2: nw.Envelope = nw.ins_envelope(wave_2)    
        
    nw.wait_for_trigger()
#     nw.wait(32e-9)
    nw.reset_frame()
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.wait(40e-9)
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.wait(80e-9)
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.wait(120e-9)
    nw.play_wave(envelope_1, 1, 0, 0)

    return nw.Kernel()
_k1 = program_1()
# device.set('ProgramOUT', _k1, f'S{slot}-O1')
for ch in range (8):
    device.set('Program', _k1, f'S{slot}-O{ch+1}')
```

```python
@nw.kernel
def program_1():
    srate: nw.Var = 4e9
    time_width_1: nw.Var = 200e-9
    freq_1: nw.Var = 0e6
    time_line_1: np.ndarray =np.linspace(0, time_width_1, int(time_width_1 * srate),endpoint=False)
    wave_1: np.ndarray = np.cos(2 * np.pi * time_line_1 * freq_1)

    time_width_2: nw.Var = 100e-9
    freq_2: nw.Var = 0e9
    time_line_2: np.ndarray =np.linspace(0, time_width_2, int(time_width_2 * srate),endpoint=False)
    wave_2: np.ndarray = -1*np.cos(2 * np.pi * time_line_2 * freq_2)
    
    frame_0: nw.Frame = nw.init_frame(0, 0.5*np.pi)
    envelope_1: nw.Envelope = nw.ins_envelope(wave_1)
    envelope_2: nw.Envelope = nw.ins_envelope(wave_2)    
        
    nw.wait_for_trigger()
#     nw.wait(32e-9)
    nw.reset_frame()
    nw.play_wave(envelope_2, 1, 0, 0)
    nw.wait(24e-9)
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.wait(60e-9)
    nw.play_wave(envelope_1, 1, 0, 0)
    nw.play_wave(envelope_2, 1, 0, 0)
    nw.play_wave(envelope_1, 1, 0, 0)

    return nw.Kernel()
_k1 = program_1()
# device.set('ProgramOUT', _k1, f'S{slot}-O1')
for ch in range (8):
    device.set('Program', _k1, f'S{slot}-O{ch+1}')
```

```python
slot =13
port_name= f'S{slot}-O1'
device.set('CustomALite', [0x01F1_0000+61*4, 0xff], port_name)
```

```python
@nw.kernel
def program_xy():
    judge_reg: nw.Reg = 0
    srate: nw.Var = 4e9
    time_width: nw.Var = 1e-6
    freq: nw.Var = 10e6
    time_line: np.ndarray = np.linspace(0, time_width, int(time_width * srate))
    wave: np.ndarray = np.cos(2 * np.pi * time_line * freq)
    frame_0: nw.Frame = nw.init_frame(0, 0.5*np.pi)
    envelope_0: nw.Envelope = nw.ins_envelope(wave)

    nw.wait_for_trigger()
    nw.play_wave(envelope_0, 1, 0, 0)
    nw.wait(2.4e-6)
    judge_reg = nw.receive(4)  # 获取广播的判决结果并存入寄存器
    if (judge_reg == 0xff): # 判决
        nw.play_wave(envelope_0, 1, 0, 0)
        nw.wait(1.2e-6)
        nw.play_wave(envelope_0, 1, 0, 0)
    else:
        nw.play_wave(envelope_0, 1, 0, 0)
    return nw.Kernel()
for ch in range(8):
    device.set('Program', program_xy(), f'S13-O{ch + 1}')
```

```python
# 10769094.-11508196.j
@nw.kernel
def program_rd():
    judge_reg0: nw.Reg = 0   
    # judge_reg1: nw.Reg = 0  # 声明两个个寄存器并赋初值
    nw.set_judge(1+0j, -10000)  # 旋转因子，判决门限，频点选择 # real[(I+jQ)(1+j0)]
    # nw.set_judge(0+1j,, [1,2,3])  # 旋转因子，判决门限，频点选择
    nw.wait_for_trigger()
    nw.wait(2.2e-6)
    judge_reg0 = nw.capture(960e-9, 160e-9, 960e-9)  # 采集判决结果并赋值给寄存器0
    nw.send(judge_reg0)

    return nw.Kernel()

# for i in range(4):
device.set('Program', program_rd(), 'S2-O4')
```

```python
device.set('TimeWidth', 960e-9, 'S2-I4')
shots = 1024
# shots = 0xffffff00
qsync.set('Shot', shots)
device.set('StartCapture')
qsync.set('GenerateTrig', 40e-6)
```
